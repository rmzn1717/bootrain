1. Plot a bar chart for total cases of the 20 countries that havebiggest numbers.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('owid-covid-data.csv', parse_dates = ["date"], low_memory = False)

df_20 = df.groupby('location').total_cases.sum().sort_values(ascending = False).head(21)

title_style = {'family': 'Arial', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Arial', 'color': 'darkblue', 'size': 15 }

plt.figure(figsize=(20, 8))
plt.bar(df_20.index, df_20)

plt.xticks(rotation = 90, fontsize = 15)

plt.xlabel('Countries',fontdict = axis_style)
plt.ylabel('Total Cases', fontdict = axis_style)

plt.title('Total Cases of 20 Countries', fontdict = title_style)

plt.show()
```

2. Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
plt.figure(figsize=(15, 7))

plt.subplot(1,3,1)
plt.hist(df[df.location == 'Germany'].new_deaths, bins = 20 )
plt.title('Daily Deaths for Germany', fontdict = title_style)
plt.xlabel('Daily Death Numbers', fontdict = axis_style)
plt.ylabel('Sum of Daily Deaths', fontdict = axis_style)

plt.subplot(1,3,2)
plt.hist(df[df.location == 'Germany'].new_deaths, bins = 50 )
plt.title('Daily Deaths for Germany', fontdict = title_style)
plt.xlabel('Daily Death Numbers', fontdict = axis_style)
plt.ylabel('Sum of Daily Deaths', fontdict = axis_style)

plt.subplot(1,3,3)
plt.hist(df[df.location == 'Germany'].new_deaths, bins = 100 )
plt.title('Daily Deaths for Germany', fontdict = title_style)
plt.xlabel('Daily Death Numbers', fontdict = axis_style)
plt.ylabel('Sum of Daily Deaths', fontdict = axis_style)

plt.show()
```

3. Plot a scatter plot of new cases and new death for Germany and France.

```python
plt.figure(figsize=(10, 6))
plt.title('New Cases & New Death', fontdict = title_style)
for location in ['Germany', 'France']:
    plt.scatter(df[df.location == location].new_cases, 
                df[df.location == location].new_deaths, label = location)
plt.xlabel('New Cases',fontdict = axis_style)
plt.ylabel('New Deaths', fontdict = axis_style)
plt.legend()
plt.show()
```

4. Plot a boxplot for daily deaths for any country you choose.

```python
plt.boxplot(df[df['location']=='Turkey'].new_deaths.dropna())
plt.title('Daily Deaths for Turkey',fontdict=title_style)
plt.ylabel('sum of daily deaths', fontdict=axis_style)
plt.xlabel('Turkey', fontdict=axis_style)
plt.show()
```

5. Calculate the total case for each continent and plot a pie chart

```python
data=df.groupby(by=['continent'])['total_cases'].sum()

plt.figure(figsize=(10, 6))
plt.title('Total Case', fontdict = title_style)

plt.pie(data, labels=data.index, autopct='%1.1f%%',
        shadow=True, startangle=45)
plt.show()
```
