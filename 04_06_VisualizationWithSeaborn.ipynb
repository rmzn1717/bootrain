1. Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
df.head()
```

```python
df_1 = df[df.location.isin(['Spain', 'France', 'Germany', 'Italy']) & (df.date >= '2020-04-01')]
df_1.head()

plt.figure(figsize=(9, 6), dpi = 500)
sns.lineplot(x = 'date', y= 'total_deaths', hue= 'location', data = df_1)
plt.show()
```

2. Plot a bar plot with seaborn for average death number that compares continents.

```python
df.groupby('continent').total_deaths.mean()

plt.figure(figsize=(9, 6), dpi = 500)
sns.barplot(x = 'continent', y = 'total_deaths', data = df)
plt.xticks(rotation = 90)
plt.show()

```

3. Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
plt.figure(figsize=(15, 10))

plt.subplot(2,2,1)
sns.distplot(df[df.location=='Germany'].new_deaths.dropna(), color = 'blue', bins =50, kde =False)
plt.title('New Deaths\n (bins=10, kde=False)' , c = 'red')
plt.ylabel('Counts')

plt.subplot(2,2,2)
sns.distplot(df[df.location=='Germany'].new_deaths.dropna(), color = 'blue', bins =20, kde =False)
plt.title('New Deaths\n (bins=10, kde=False)' , c = 'red')
plt.ylabel('Counts')

plt.subplot(2,2,3)
sns.distplot(df[df.location=='Germany'].new_deaths.dropna(), color = 'blue', bins =50)
plt.title('New Deaths\n (bins=10, kde=False)' , c = 'red')
plt.ylabel('Normalized Counts')

plt.subplot(2,2,4)
sns.distplot(df[df.location=='Germany'].new_deaths.dropna(), color = 'blue', bins =20)
plt.title('New Deaths\n (bins=10, kde=False)' , c = 'red')
plt.ylabel('Normalized Counts')

plt.show()
```

4. Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
df4 = df[df.location.isin(['Germany', 'Italy'])]
          
plt.figure(figsize=(15,5))
plt.subplot(131)
sns.swarmplot(x="location", y="new_deaths", data=df4)
          
plt.subplot(132)
sns.boxplot(x="location", y="new_deaths", data=df4)

plt.subplot(133)
sns.violinplot(x="location", y="new_deaths", data=df4)

plt.show()
```
